```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_buckets = 6
b_size = n_buckets - 1
probs = np.full((n_buckets,10),1/n_buckets)
bucket = np.full(b_size,1)
```

```python
for i in range(10):
    for j,prob in enumerate(probs):
        if i == 0:
            if np.random.choice(bucket):
                prob[i] = prob[i]*(j/b_size)
            else:
                prob[i] = prob[i]*((b_size-j)/b_size)
        else:
            if np.random.choice(bucket):
                prob[i] = prob[i-1]*(j/b_size)
            else:
                prob[i] = prob[i-1]*((b_size-j)/b_size)
    probs[:,i] = probs[:,i]/sum(probs[:,i])
            
        
```

```python
fig, ax = plt.subplots(2,3, figsize= (20,10))

for i,ax in enumerate(ax.flatten()):
    ax.set_title(f"H{i}")
    ax.plot(np.arange(10),probs[i], 'bo')
```
