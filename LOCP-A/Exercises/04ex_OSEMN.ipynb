1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
#Solution
#List creation
l=np.random.rand(10)
#Writing every number to the file
f= open("simple_data.txt", mode='w')
for i in l:
    f.write(str(i)+"\n")
#Checking the result
with open("simple_data.txt", mode='r') as f:
    for line in f:
        print (line)
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
#Solution
#Matrix creation
l=np.random.rand(5,5)
print(l,"\n")
#Writing the matrix to the file
f= open("data.txt", mode='w')
for i in l:
    for j in i:
        f.write(str(j)+" ")
    f.write("\n")
#Checking the result
with open("data.txt", mode='r') as f:
    for line in f:
        print (line)
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
#Solution
#Creating the file to write
csv=open("csv.data", mode='w')
#Opening the point 2 file
with open("data.txt", mode='r') as f:
    for line in f:
        for i,c in enumerate(line.split()):
            csv.write(c)
            if i!=(len(line.split())-1):
                csv.write(",")
        csv.write("\n")
#Checking the result
with open("csv.data", mode='r') as f:
    for line in f:
        print (line)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
#Solution
file="credit_card.dat"
lenght=6
with open(file, mode='rb') as f:
    for line in f:
        for j in range(0,len(line),lenght): 
            print(chr(int(line[j:j+lenght],2)),end="") #Printing the corresponding char for 
                                                       #Each binary
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
#Solution
import json
import csv
#Loading th json
data = json.load(open('user_data.json'))
data2=[]
#Writing as a CSV
file=open("user_data.data", mode='w')
writer = csv.writer(file, delimiter = ',')
writer.writerow(['ID','JobTitle','Email Address', 'FirstNameLastName','CreditCard','CreditCardType'])
for dic in data:
    if dic["CreditCardType"]=="American Express":
        data2.append(dic)
        writer.writerow(list(dic.values()))
print(data2)
#In this way I print as the first row the csv the keys of dictionary
#The remaining rows corresponds to the values associated to each key
```
